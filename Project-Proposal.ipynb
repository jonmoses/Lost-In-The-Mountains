## Exploratory Analysis

In order to visualize our data set, we will want to use exploratory data analysis which will answer questions we may want to find out. For instance, we will want to see which areas of the Appalachia have the most missing people, we will want to find out which state has the most missing people, and so then we can draw some interesting conclusions for this. In order to do this we can compare the percentages of missing people across states and certain areas within the states. We can also analyze the data spatially, meaning thar we can we use a scatterplot using latitude and logitude where each dot is a missing person's last seen location on a map. This will help us indentify patterns and visualize where the most people go missing. We can also make a density map if the clustering becomes too crowded and makes our scatter plot hard to interpret. Too better look at our dataset we can also compare the trends of missing people, meaning we can take the data last seen and see how it changes over the years. This will tell us when there is an increase or decrease if people going missing, and allow us the analyze the trends of certain time periods. We can do this using frequency counts, and then use a line graph to visualize the trends.We can also use bar charts to compare the gender of people missing or the age of people missing, and we can also use histograms to compare the age distribution of people that have gone missing

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv("data/missing_persons_with_coords.csv")

# Format the date with pandas to_datetime method, data is in m/d/y in the csv from NamUs.
df["DLC"] = pd.to_datetime(df["DLC"], format="%m/%d/%Y")
df["Year"] = df["DLC"].dt.year
df["Month"] = df["DLC"].dt.month
df["Age"] = pd.to_numeric(df["Missing Age"].str.replace(" Years", "", regex=False), errors="coerce")


# Race has multiple entries sometimes if a person is mixed so im just combining them into a new multiracial
# Category for now. We should probably change this later on in the project and handle it more carefully.
df["Race"] = df["Race / Ethnicity"].apply(lambda x: "Multiracial" if "," in str(x) else x)
```

```python
# Plotting demographic data

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Demographic Overview of Missing Persons Cases", fontsize=14, fontweight="bold")

# cound the gender column and graph it
sex_counts = df["Biological Sex"].value_counts()
axes[0].bar(sex_counts.index, sex_counts.values, color=["steelblue", "salmon"])
axes[0].set_title("Cases by Gender")
axes[0].set_ylabel("Number of Cases")
for i, v in enumerate(sex_counts.values):
    axes[0].text(i, v + 20, f"{v:,}", ha="center", fontsize=10)

# counting the race column and graphing it
race_counts = df["Race"].value_counts()
axes[1].barh(race_counts.index, race_counts.values)
axes[1].set_title("Cases by Race / Ethnicity")
axes[1].set_xlabel("Number of Cases")
axes[1].invert_yaxis()

# histogram for age. 30 bins looks pretty good.
axes[2].hist(df["Age"].dropna(), bins=30)
axes[2].set_title("Age at Disappearance")
axes[2].set_xlabel("Age (Years)")
axes[2].set_ylabel("Number of Cases")

plt.tight_layout()
plt.show()
```

```python
# Plotting cases over time

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle("Missing Persons Cases Over Time", fontsize=14, fontweight="bold")

year_counts = df.groupby("Year").size()

axes[0].plot(year_counts.index, year_counts.values, marker="o")
axes[0].set_title("Cases per Year (Date Last Contact)")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Number of Cases")
axes[0].xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

#
month_counts = df.groupby("Month").size()
month_labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
axes[1].bar(month_counts.index, month_counts.values)
axes[1].set_title("Cases by Month")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Number of Cases")
axes[1].set_xticks(range(1, 13))
axes[1].set_xticklabels(month_labels)

plt.tight_layout()
plt.show()
```

## Analysis Methodolgy

Our analysis is organized into four main areas: geospatial, demographic, temporal, and economic. Each section below describes the specific methods and visualizations we plan to use.



#### Geospatial Analysis
The core of our analysis is geographic. We will use two approaches.

First, we will produce heatmaps of the Appalachian region showing the count of missing persons cases per county across the 18-states in the area. The result will be an interactive map where color intensity corresponds to case count, making regional concentrations immediately visible.

Second, we will apply K-means clustering following the methods introduced in Lecture 19: Clustering 1 to identify spatial hotspots at the point level, using the county coordinates attached to each case. The output will be visualized as a scatter plot overlaid on a map, with each cluster colored distinctly.



#### Demographic Analysis

#### Economic Analysis
